```python
import os, re
import random, math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
# 데이터 로드할 때 빠르게 로드할 수 있도록하는 설정 변수
AUTOTUNE = tf.data.experimental.AUTOTUNE
# X-RAY 이미지 사이즈 변수
IMAGE_SIZE = [180, 180]

# 데이터 경로 변수
TRAIN_PATH =  'data/train/*/*' # *은 모든 디렉토리와 파일을 의미합니다.
VAL_PATH =  'data/val/*/*'
TEST_PATH =  'data/test/*/*'

```

## 실험환경 Set-up

```python
# 프로젝트를 진행할 때 아래 두 변수를 변경해보세요
BATCH_SIZE = 16
EPOCHS = 10
```

# ablation study를 수행하기 위한 기준모델 설정 

```python

# Define the model class
class PneumoniaDetectionModel:
    def __init__(self):
        # 데이터 로딩 함수 호출
        self.load_and_prepare_data()
        # 모델 생성 및 컴파일 로직 
        self.model = self.build_model()
        self.compile_model()
        self.is_show_batch = False
    def load_and_prepare_data(self):
        # 데이터 파일 경로 불러오기
        self.train_filenames = tf.io.gfile.glob(TRAIN_PATH)
        self.val_filenames = tf.io.gfile.glob(VAL_PATH)
        self.test_filenames = tf.io.gfile.glob(TEST_PATH)

        print("------------------훈련,테스트,검증 데이터 개수-------------------")
        print("train_filenames 개수: ", len(self.train_filenames))
        print("test_filenames 개수: ", len(self.test_filenames))
        print("val_filenames 개수: ",len(self.val_filenames))
        print("-----------------------------------------")


        # train 데이터와 validation 데이터를 모두 filenames에 담습니다
        self.filenames = tf.io.gfile.glob(TRAIN_PATH)
        self.filenames.extend(tf.io.gfile.glob(VAL_PATH))

        # 모아진 filenames를 8:2로 나눕니다
        self.train_size = math.floor(len(self.filenames)*0.8)
        random.seed(8)
        random.shuffle(self.filenames)
        self.train_filenames = self.filenames[:self.train_size]
        self.val_filenames = self.filenames[self.train_size:]
        print("------------------훈련,검증 데이터 개수-------------------")
        print(len(self.train_filenames))
        print(len(self.val_filenames))
        print("-----------------------------------------------------")


        self.COUNT_NORMAL = len([filename for filename in self.train_filenames if "NORMAL" in filename])


        self.COUNT_PNEUMONIA = len([filename for filename in self.train_filenames if "PNEUMONIA" in filename])
        print("------------------클래스당 개수-------------------")
        print(f"Normal images count in training set: {self.COUNT_NORMAL}")
        print(f"Pneumonia images count in training set: {self.COUNT_PNEUMONIA}")
        print("-----------------------------------------")

        # TensorFlow 데이터셋 객체 생성
        self.train_list_ds = tf.data.Dataset.from_tensor_slices(self.train_filenames)
        self.val_list_ds = tf.data.Dataset.from_tensor_slices(self.val_filenames)
        self.test_list_ds = tf.data.Dataset.list_files(TEST_PATH)
        # 이미지 수 계산
        self.TRAIN_IMG_COUNT = tf.data.experimental.cardinality(self.train_list_ds).numpy()


        self.VAL_IMG_COUNT = tf.data.experimental.cardinality(self.val_list_ds).numpy()
        self.TEST_IMG_COUNT = tf.data.experimental.cardinality(self.test_list_ds).numpy()
        print("------------------이미지 개수-------------------")
        print(f"Training images count: {self.TRAIN_IMG_COUNT}")
        print(f"Validating images count: {self.VAL_IMG_COUNT}")
        print(f"Testing images count: {self.TEST_IMG_COUNT}")
        print("-----------------------------------------")


    @staticmethod
    def show_batch( image_batch, label_batch):
        plt.figure(figsize=(10,10))
        for n in range(BATCH_SIZE):
            ax = plt.subplot(4,math.ceil(BATCH_SIZE/4),n+1)
            plt.imshow(image_batch[n])
            if label_batch[n]:
                plt.title("PNEUMONIA")
            else:
                plt.title("NORMAL")
            plt.axis("off")


    @staticmethod
    def get_label(file_path):
        parts = tf.strings.split(file_path, os.path.sep)
        label = parts[-2] == "PNEUMONIA"
        return tf.cast(label, tf.float32)  # bool을 float32로 변환

    @staticmethod
    def decode_img(img):
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.convert_image_dtype(img, tf.float32)
        img = tf.image.resize(img, IMAGE_SIZE)
        return img
    def process_path(self, file_path):
        label = self.get_label(file_path)
        img = tf.io.read_file(file_path)
        img = self.decode_img(img)
        return img, label
    @staticmethod
    def prepare_for_training( ds, shuffle_buffer_size=1000):
        ds = ds.shuffle(buffer_size=shuffle_buffer_size)
        ds = ds.repeat()
        ds = ds.batch(BATCH_SIZE)
        ds = ds.prefetch(buffer_size=AUTOTUNE)
        return ds
    @staticmethod
    def conv_block(filters):
        block = tf.keras.Sequential([
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.SeparableConv2D(filters, 3, activation='relu', padding='same'),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.MaxPool2D()
        ])
        
        return block
    @staticmethod
    def dense_block(units, dropout_rate):
        block = tf.keras.Sequential([
            tf.keras.layers.Dense(units, activation='relu'),
            tf.keras.layers.BatchNormalization(),
            tf.keras.layers.Dropout(dropout_rate)
        ])
        
        return block

    def build_model(self):
        model = tf.keras.Sequential([
            tf.keras.Input(shape=(IMAGE_SIZE[0], IMAGE_SIZE[1], 3)),
            
            tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
            tf.keras.layers.Conv2D(16, 3, activation='relu', padding='same'),
            tf.keras.layers.MaxPool2D(),
            
            self.conv_block(32),
            self.conv_block(64),
            
            self.conv_block(128),
            tf.keras.layers.Dropout(0.2),
            
            self.conv_block(256),
            tf.keras.layers.Dropout(0.2),
            
            tf.keras.layers.Flatten(),
            self.dense_block(512, 0.7),
            self.dense_block(128, 0.5),
            self.dense_block(64, 0.3),
            
            tf.keras.layers.Dense(1, activation='sigmoid')
        ])
        
        return model

    def compile_model(self):
        METRICS = [
            'accuracy',
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
        
        self.model.compile(
            optimizer='adam',
            loss='binary_crossentropy',
            metrics=METRICS
        )

    def train(self):
        # 데이터셋 전처리
        train_ds = self.train_list_ds.map(self.process_path, num_parallel_calls=AUTOTUNE)
        val_ds = self.val_list_ds.map(self.process_path, num_parallel_calls=AUTOTUNE)

        print("------------------샘플 한개 뽑아보기-------------------")
        for image, label in train_ds.take(1):
            print("Image shape: ", image.numpy().shape)
            print("Label: ", label.numpy())
        print("-----------------------------------------------------")

        train_ds = self.prepare_for_training(train_ds)
        val_ds = self.prepare_for_training(val_ds)
        image_batch, label_batch = next(iter(train_ds))
        if self.is_show_batch:
            self.show_batch(image_batch.numpy(), label_batch.numpy())
        # 가중치 계산
        weight_for_0 = (1 / self.COUNT_NORMAL)*(self.TRAIN_IMG_COUNT)/2.0 
        weight_for_1 = (1 / self.COUNT_PNEUMONIA)*(self.TRAIN_IMG_COUNT)/2.0

        class_weight = {0: weight_for_0, 1: weight_for_1}
        print("------------------가중치 계산-------------------")
        print('Weight for NORMAL: {:.2f}'.format(weight_for_0))
        print('Weight for PNEUMONIA: {:.2f}'.format(weight_for_1))
        print("-----------------------------------------")


        # 모델 훈련
        history = self.model.fit(
            train_ds,
            epochs=EPOCHS,
            validation_data=val_ds,
            steps_per_epoch=self.TRAIN_IMG_COUNT // BATCH_SIZE,
            validation_steps=self.VAL_IMG_COUNT // BATCH_SIZE,
            class_weight=class_weight
        )
        return history
    # 테스트 데이터에 대한 모델 평가
    def evaluate(self):
        self.test_ds = self.test_list_ds.map(self.process_path, num_parallel_calls=AUTOTUNE)
        self.test_ds = self.test_ds.batch(BATCH_SIZE)
        self.test_ds = self.test_ds.cache().prefetch(buffer_size=AUTOTUNE)
        loss, accuracy, precision, recall = self.model.evaluate(self.test_ds)
        print("------------------성능지표-------------------")
        print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
        print("-----------------------------------------")
        return loss, accuracy, precision, recall 
    def show_graph(self, history):
        fig, ax = plt.subplots(1, 4, figsize=(20, 3))
        ax = ax.ravel()

        for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
            ax[i].plot(history.history[met])
            ax[i].plot(history.history['val_' + met])
            ax[i].set_title('Model {}'.format(met))
            ax[i].set_xlabel('epochs')
            ax[i].set_ylabel(met)
            ax[i].legend(['train', 'val'])
       

```

```python
# PneumoniaDetectionModel 클래스의 인스턴스 생성
model = PneumoniaDetectionModel()
# 모델 훈련
history = model.train()
# 훈련 과정의 결과 (정확도, 손실 등) 시각화
model.show_graph(history)
# 테스트 데이터에 대한 모델 평가
test_results = model.evaluate()
```

## augmentation 적용여부 , regularization 변수로 처리해서 ablation study 준비

```python

# Define the model class
class PneumoniaDetectionModel2:
    def __init__(self):
        # 데이터 로딩 함수 호출
        self.model=None
        self.load_and_prepare_data()
        self.is_show_batch=False
        # 모델 생성 및 컴파일 로직 
    def set_model(self,model):
        self.model=model
    def load_and_prepare_data(self):
        # 데이터 파일 경로 불러오기
        self.train_filenames = tf.io.gfile.glob(TRAIN_PATH)
        self.val_filenames = tf.io.gfile.glob(VAL_PATH)
        self.test_filenames = tf.io.gfile.glob(TEST_PATH)

        print("------------------훈련,테스트,검증 데이터 개수-------------------")
        print("train_filenames 개수: ", len(self.train_filenames))
        print("test_filenames 개수: ", len(self.test_filenames))
        print("val_filenames 개수: ",len(self.val_filenames))
        print("-----------------------------------------")


        # train 데이터와 validation 데이터를 모두 filenames에 담습니다
        self.filenames = tf.io.gfile.glob(TRAIN_PATH)
        self.filenames.extend(tf.io.gfile.glob(VAL_PATH))

        # 모아진 filenames를 8:2로 나눕니다
        self.train_size = math.floor(len(self.filenames)*0.8)
        random.seed(8)
        random.shuffle(self.filenames)
        self.train_filenames = self.filenames[:self.train_size]
        self.val_filenames = self.filenames[self.train_size:]
        print("------------------훈련,검증 데이터 개수-------------------")
        print(len(self.train_filenames))
        print(len(self.val_filenames))
        print("-----------------------------------------------------")


        self.COUNT_NORMAL = len([filename for filename in self.train_filenames if "NORMAL" in filename])


        self.COUNT_PNEUMONIA = len([filename for filename in self.train_filenames if "PNEUMONIA" in filename])
        print("------------------클래스당 개수-------------------")
        print(f"Normal images count in training set: {self.COUNT_NORMAL}")
        print(f"Pneumonia images count in training set: {self.COUNT_PNEUMONIA}")
        print("-----------------------------------------")

        # TensorFlow 데이터셋 객체 생성
        self.train_list_ds = tf.data.Dataset.from_tensor_slices(self.train_filenames)
        self.val_list_ds = tf.data.Dataset.from_tensor_slices(self.val_filenames)
        self.test_list_ds = tf.data.Dataset.list_files(TEST_PATH)
        # 이미지 수 계산
        self.TRAIN_IMG_COUNT = tf.data.experimental.cardinality(self.train_list_ds).numpy()


        self.VAL_IMG_COUNT = tf.data.experimental.cardinality(self.val_list_ds).numpy()
        self.TEST_IMG_COUNT = tf.data.experimental.cardinality(self.test_list_ds).numpy()
        print("------------------이미지 개수-------------------")
        print(f"Training images count: {self.TRAIN_IMG_COUNT}")
        print(f"Validating images count: {self.VAL_IMG_COUNT}")
        print(f"Testing images count: {self.TEST_IMG_COUNT}")
        print("-----------------------------------------")


    @staticmethod
    def show_batch( image_batch, label_batch):
        plt.figure(figsize=(10,10))
        for n in range(BATCH_SIZE):
            ax = plt.subplot(4,math.ceil(BATCH_SIZE/4),n+1)
            plt.imshow(image_batch[n])
            if label_batch[n]:
                plt.title("PNEUMONIA")
            else:
                plt.title("NORMAL")
            plt.axis("off")


    @staticmethod
    def get_label(file_path):
        parts = tf.strings.split(file_path, os.path.sep)
        label = parts[-2] == "PNEUMONIA"
        return tf.cast(label, tf.float32)  # bool을 float32로 변환

    @staticmethod
    def decode_img(img):
        img = tf.image.decode_jpeg(img, channels=3)
        img = tf.image.convert_image_dtype(img, tf.float32)
        img = tf.image.resize(img, IMAGE_SIZE)
        return img
    def process_path(self, file_path):
        label = self.get_label(file_path)
        img = tf.io.read_file(file_path)
        img = self.decode_img(img)
        return img, label
    @staticmethod
    def prepare_for_training( ds, shuffle_buffer_size=1000):
        ds = ds.shuffle(buffer_size=shuffle_buffer_size)
        ds = ds.repeat()
        ds = ds.batch(BATCH_SIZE)
        ds = ds.prefetch(buffer_size=AUTOTUNE)
        return ds
    
    def compile_model(self,optimizer='adam'):
        # print("------------------모델 컴파일-------------------")
        # print(f'optimizer: {optimizer}')
        # print("-----------------------------------------")
        METRICS = [
            'accuracy',
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
        
        self.model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=METRICS)

    def train(self, augment=False):
        # 데이터셋 전처리
        train_ds = self.train_list_ds.map(self.process_path, num_parallel_calls=AUTOTUNE)
        val_ds = self.val_list_ds.map(self.process_path, num_parallel_calls=AUTOTUNE)
        if augment:
            # 데이터 증강 로직 추가
            train_ds = train_ds.map(self.augment, num_parallel_calls=AUTOTUNE)
        # print("------------------샘플 한개 뽑아보기-------------------")
        # for image, label in train_ds.take(1):
        #     print("Image shape: ", image.numpy().shape)
        #     print("Label: ", label.numpy())
        # print("-----------------------------------------------------")

        train_ds = self.prepare_for_training(train_ds)
        val_ds = self.prepare_for_training(val_ds)
        image_batch, label_batch = next(iter(train_ds))
        if self.is_show_batch:
            self.show_batch(image_batch.numpy(), label_batch.numpy())
        # 가중치 계산
        weight_for_0 = (1 / self.COUNT_NORMAL)*(self.TRAIN_IMG_COUNT)/2.0 
        weight_for_1 = (1 / self.COUNT_PNEUMONIA)*(self.TRAIN_IMG_COUNT)/2.0

        class_weight = {0: weight_for_0, 1: weight_for_1}
        print("------------------가중치 계산-------------------")
        print('Weight for NORMAL: {:.2f}'.format(weight_for_0))
        print('Weight for PNEUMONIA: {:.2f}'.format(weight_for_1))
        print("-----------------------------------------")


        # 모델 훈련
        history = self.model.fit(
            train_ds,
            epochs=EPOCHS,
            validation_data=val_ds,
            steps_per_epoch=self.TRAIN_IMG_COUNT // BATCH_SIZE,
            validation_steps=self.VAL_IMG_COUNT // BATCH_SIZE,
            class_weight=class_weight
        )
        return history
    # 테스트 데이터에 대한 모델 평가
    def evaluate(self):
        test_ds = self.test_list_ds.map(self.process_path, num_parallel_calls=AUTOTUNE)
        test_ds = test_ds.batch(BATCH_SIZE)
        test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)
        loss, accuracy, precision, recall = self.model.evaluate(test_ds)
        print("------------------성능지표-------------------")
        print(f'Loss: {loss},\nAccuracy: {accuracy},\nPrecision: {precision},\nRecall: {recall}')
        print("-----------------------------------------")
        return loss, accuracy, precision, recall 
    def show_graph(self, history):
        fig, ax = plt.subplots(1, 4, figsize=(20, 3))
        ax = ax.ravel()

        for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
            ax[i].plot(history.history[met])
            ax[i].plot(history.history['val_' + met])
            ax[i].set_title('Model {}'.format(met))
            ax[i].set_xlabel('epochs')
            ax[i].set_ylabel(met)
            ax[i].legend(['train', 'val'])
    def show_graph2(self, history, condition):
        fig, ax = plt.subplots(1, 4, figsize=(24, 4))  # 그래프 크기 조절
        ax = ax.ravel()

        for i, met in enumerate(['precision', 'recall', 'accuracy', 'loss']):
            ax[i].plot(history.history[met])
            ax[i].plot(history.history['val_' + met])
            ax[i].set_title(f'Model {met} -\n{condition}', fontsize=10)  # 줄바꿈 및 글꼴 크기 조절
            ax[i].set_xlabel('epochs')
            ax[i].set_ylabel(met)
            ax[i].legend(['train', 'val'])

        plt.subplots_adjust(wspace=0.4, hspace=0.6)  # 서브플롯 간격 조절
        plt.show()

    @staticmethod
    def augment(image,label):
        image = tf.image.random_flip_left_right(image)  # 랜덤하게 좌우를 반전합니다.
        return image,label

```

```python
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, BatchNormalization, Activation, Add, MaxPooling2D, GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model

class ResNet18(Model):
    def __init__(self, input_shape, num_classes):
        super(ResNet18, self).__init__()
        self.conv1 = Conv2D(64, (7, 7), strides=(2, 2), padding='same', input_shape=input_shape)
        self.bn1 = BatchNormalization()
        self.relu = Activation('relu')
        self.max_pool = MaxPooling2D((3, 3), strides=(2, 2), padding='same')

        # Define the residual blocks
        self.res_block1 = self.build_res_block(64, 2, first_block=True)
        self.res_block2 = self.build_res_block(128, 2)
        self.res_block3 = self.build_res_block(256, 2)
        self.res_block4 = self.build_res_block(512, 2)

        # Output layers
        self.avg_pool = GlobalAveragePooling2D()
        self.fc = Dense(num_classes, activation='softmax')

    def call(self, inputs):
        x = self.conv1(inputs)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.max_pool(x)

        # Apply residual blocks
        x = self.res_block1(x)
        x = self.res_block2(x)
        x = self.res_block3(x)
        x = self.res_block4(x)

        # Output layers
        x = self.avg_pool(x)
        return self.fc(x)

    def build_res_block(self, filters, blocks, stride=1, first_block=False):
        res_blocks = tf.keras.Sequential()
        for i in range(blocks):
            if i == 0 and not first_block:
                res_blocks.add(ResidualBlock(filters, stride=stride, downsample=True))
            else:
                res_blocks.add(ResidualBlock(filters, stride=1, downsample=False))
        return res_blocks

class ResidualBlock(Model):
    def __init__(self, filters, stride=1, downsample=False):
        super(ResidualBlock, self).__init__()
        self.conv1 = Conv2D(filters, (3, 3), strides=stride, padding='same')
        self.bn1 = BatchNormalization()
        self.relu = Activation('relu')
        self.conv2 = Conv2D(filters, (3, 3), strides=1, padding='same')
        self.bn2 = BatchNormalization()
        self.downsample = downsample
        if downsample:
            self.conv_downsample = Conv2D(filters, (1, 1), strides=stride)

    def call(self, inputs):
        residual = inputs
        x = self.conv1(inputs)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.conv2(x)
        x = self.bn2(x)

        if self.downsample:
            residual = self.conv_downsample(residual)

        x = Add()([x, residual])
        return self.relu(x)

# Example usage
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Conv2D, MaxPooling2D, Dropout, Flatten, BatchNormalization, SeparableConv2D

class CustomModel(Model):
    def __init__(self, input_shape, dropout_rate=0.2, regularization=None):
        super(CustomModel, self).__init__()
        self.regularization = regularization
        self.dropout_rate = dropout_rate

        # Convolutional layers
        self.conv1 = Conv2D(16, 3, activation='relu', padding='same', kernel_regularizer=regularization, input_shape=input_shape)
        self.conv2 = Conv2D(16, 3, activation='relu', padding='same', kernel_regularizer=regularization)
        self.max_pool = MaxPooling2D()

        # Separable Convolutional Blocks
        self.sep_conv_block1 = self.create_sep_conv_block(32)
        self.sep_conv_block2 = self.create_sep_conv_block(64)
        self.sep_conv_block3 = self.create_sep_conv_block(128)
        self.dropout1 = Dropout(dropout_rate)
        self.sep_conv_block4 = self.create_sep_conv_block(256)
        self.dropout2 = Dropout(dropout_rate)

        # Dense blocks
        self.flatten = Flatten()
        self.dense_block1 = self.create_dense_block(512)
        self.dense_block2 = self.create_dense_block(128)
        self.dense_block3 = self.create_dense_block(64)

        # Output layer
        self.output_layer = Dense(1, activation='sigmoid')

    def call(self, inputs):
        x = self.conv1(inputs)
        x = self.conv2(x)
        x = self.max_pool(x)

        x = self.sep_conv_block1(x)
        x = self.sep_conv_block2(x)
        x = self.sep_conv_block3(x)
        x = self.dropout1(x)
        x = self.sep_conv_block4(x)
        x = self.dropout2(x)

        x = self.flatten(x)
        x = self.dense_block1(x)
        x = self.dense_block2(x)
        x = self.dense_block3(x)

        return self.output_layer(x)

    def create_sep_conv_block(self, filters):
        block = tf.keras.Sequential([
            SeparableConv2D(filters, 3, activation='relu', padding='same', kernel_regularizer=self.regularization),
            SeparableConv2D(filters, 3, activation='relu', padding='same', kernel_regularizer=self.regularization),
            BatchNormalization(),
            MaxPooling2D()
        ])
        return block

    def create_dense_block(self, units):
        block = tf.keras.Sequential([
            Dense(units, activation='relu'),
            BatchNormalization(),
            Dropout(self.dropout_rate)
        ])
        return block

```

```python
def run_experiments(workflow, input_shape, learning_rates, augment_conditions, dropout_rates, regularization_strengths, optimizers ,is_show_batch=False):
    results = {}
    workflow.is_show_batch=is_show_batch 
    for lr in learning_rates:
        for augment in augment_conditions:
            for dropout_rate in dropout_rates:
                for optimizer in optimizers:
                    for reg_strength in regularization_strengths:
                        print("------------------새로운 실험-------------------")

                        print(f"Testing with augment: {augment}, dropout rate: {dropout_rate}, regularization strength: {reg_strength} , optimizer: {optimizer} , learning rate: {lr}")
                        
                        # 모델 생성
                        custom_model = CustomModel(input_shape, dropout_rate=dropout_rate, regularization=tf.keras.regularizers.l2(reg_strength))
                        workflow.model = custom_model
                        if(optimizer=='adam'):
                            optimizer2 = tf.keras.optimizers.Adam(learning_rate=lr)
                            workflow.compile_model(optimizer=optimizer2)
                        else:
                            optimizer2 = tf.keras.optimizers.RMSprop(learning_rate=lr)
                            workflow.compile_model(optimizer=optimizer2)
                        # 모델 훈련 (여기서는 예시로 1 epoch만 진행)
                        history = workflow.train(augment=augment)
                        if(optimizer=='adam'):
                            workflow.show_graph(history)
                        else:
                            workflow.show_graph(history)
                        workflow.evaluate()

```
